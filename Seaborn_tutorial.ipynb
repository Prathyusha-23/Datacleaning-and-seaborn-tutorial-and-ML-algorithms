```python
#subplots in seaborn(one library for plotting,more complex graphs are also drawn than matplotlib)
#FacetGrid is used for subplots
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","tip")
plt.show()

g=sb.FacetGrid(df,col="time",hue="sex")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
#stastical estimation
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
#Linear relationship(regrssion plot)
#regression plot is used for predicting the future without having the value
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
#pair plot
sb.pairplot(df)
plt.show()
```

```python
#pairplots with selected parameters
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
#adding a hue color to a seaborn pairplot
sb.pairplot(df,hue='size')
plt.show()

sb.pairplot(df,hue='size',palette="husl")
plt.show()
```

```python
sb.pairplot(df,diag_kind="kde")
plt.show()
```
